```python
from collections import Counter

# Calculate the frequency count
category_0 = Counter(df_0['Department'])
category_1 = Counter(df_1['Department'])

# Extract categories and counts from the Counter object
categories_0 = list(category_0.keys())
counts_0 = list(category_0.values())

xpos = np.array(np.arange(len(categories_0)))

categories_1 =list(category_1.keys())
counts_1 = list(category_1.values())

# Create the bar chart
plt.bar(xpos-.2, counts_0, width = .4)
plt.bar(xpos+.2, counts_1, width = .4)

plt.xticks(xpos, categories_0, rotation = 'vertical')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('Bar Chart')

# Display the chart
plt.show()
```

```python
merged['salary'] = salary
```

```python
final = merged.drop('Department', axis = 1)
```

```python
final.head()
```

```python
from sklearn.model_selection import train_test_split
X = final.drop('left', axis = 1)
y = final['left']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
