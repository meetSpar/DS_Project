```python
X = np.stack([np.ones(m), X], axis=1)
theta = np.array([0, 0])
```

```python
def computeCost(X, y, theta):
    m = y.size
    b = np.dot(X, theta)-y
    J = 0
    
    J = np.sum(b**2)
    return (0.5/m)*J
```

```python
computeCost(X,y,theta)
```

```python
computeCost(X,y,[-1, 2])
```

```python
def gradientDesc(X, y, theta, alpha, n_iters):
    theta = theta.copy()
    J_history = []
    for i in range(n_iters):
        theta = theta-alpha*np.dot(X.T, (np.dot(X, theta)-y))/m
        J_history.append(computeCost(X,y,theta))
        
    return theta, J_history
```

```python
theta_f, J_history = gradientDesc(X, y, theta, 0.01, 1500)
```

```python
theta_f
```

```python
plt.scatter(X[:,1], y, marker='o', c='b')
plt.plot(X[:,1], np.dot(X,theta_f))
```

```python
computeCost(X, y, theta_f)
```

```python
def predict(X):
    return np.dot(X, theta_f)
```
