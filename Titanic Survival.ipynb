```python
df1 = df.drop(['PassengerId', 'Name', 'SibSp',
       'Parch', 'Ticket', 'Cabin', 'Embarked'], axis=1)
```

```python
df1
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le_sex = le.fit_transform(df1['Sex'])
```

```python
df2 = df1.copy()
```

```python
df2['Sex'] = le_sex
```

```python
df3 = df2.dropna()
```

```python
df3
```

```python
X = df3.drop('Survived', axis=1)
y = df3.Survived
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

lr = LogisticRegression()
sv = SVC()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.2, random_state=10)

lr.fit(X_train, y_train)
sv.fit(X_train, y_train)
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import ShuffleSplit
ss = ShuffleSplit(n_splits = 5, test_size = .2, random_state = 0)

from sklearn.model_selection import cross_val_score
```

```python
import numpy as np
model = [lr, sv, dt, rf]

for m in model:
    score = cross_val_score(m, X, y, cv=ss)
    print(np.mean(score))
```

```python
def predict_survival(Pclass, Sex, Age, Fare):
    x = np.zeros(len(X.columns))
    x[0] = Pclass
    x[2] = Age
    x[3] = Fare
    
    if Sex.lower()=='male':
        x[1] = 1
    
    return rf.predict([x])[0]
    
```
